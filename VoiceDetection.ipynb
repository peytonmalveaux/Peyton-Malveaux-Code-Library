```python
import numpy as np
import pandas as pd

#splits data into training and testing automatically
from sklearn.model_selection import train_test_split

#import logestic regression machine learning model for simple binary classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

```

Data Collection

```python
#load dataset into a pandas dataset

voice_data = pd.read_csv('/voice.csv')
```

```python
voice_data.head()
```

```python
#gives us the numbers of rows and columns(BINARY DATA IS ON THE LAST COLUMN)
voice_data.shape
```

```python
voice_data.describe()
```

```python
voice_data['label'] = voice_data['label'].astype(str)
voice_data['label'].value_counts()

```

```python
voice_data.groupby('label').mean()
```

```python
#seperate data and label
X = voice_data.drop(columns='label', axis=1)
Y = voice_data['label']

#label data goes into Y, everything else goes into X
print(X)
print(Y)
```

TRAIN AND TEST DATA

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size=.2, stratify = Y, random_state=1)

print(X.shape, xtrain.shape, xtest.shape)
```

```python
print(xtrain, ytrain)
```

Logestic Regression Model Training

```python
#model = LogisticRegression()
model = LogisticRegression(max_iter=5000)
#training the model w training data
model.fit(xtrain, ytrain)
```

Model Evaluation

```python
#accuracy on training data

xtrain_prediction = model.predict(xtrain)
training_accuracy = accuracy_score(xtrain_prediction, ytrain)
print("Accuracy: ", training_accuracy)
```
